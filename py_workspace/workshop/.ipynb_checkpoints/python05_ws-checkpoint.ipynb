# 방카슈랑스 이율 분석하기

```python
with open("../data/방카슈랑스.csv", 'r', ) as file:
    banca = file.read()
banca
```

### 읽어온 방카슈랑스 csv 파일을 split 함수를 사용해서 다음과 같이 분리해 보세요.

```python
banca=banca.split("\n")[1:]
banca
```

## 위 list의 각 요소에 다시 split을 사용해 보험사, 상품명, 공시이율로 나누어 보세요.

```python
sList=[]
for i in range(0,len(banca)):
    bList=[]
    temp=banca[i].split(",")
    for i in range(0,4):
        if i==2:
            pass
        else:
            bList.append(temp[i])
    sList.append(",".join(bList))
print(sList)
        
```

## 함수 bestBanc(comp)를 다음과 같이 만들어보세요.
### 한 보험사의 가장 이율이 높은 보험상품을 찾아주는 함수입니다.
1. 매개변수 comp는 가입할 보험사의 이름입니다.
2. 입력받은 보험사의 보험 중 이율이 가장 높은 것을 찾아 보험의 이름과 이율을 return 해주세요.

```python
def bestBanc(comp):
    max_comp=[]
    for i in sList:
        if comp in i:
               max_comp.append(i[-4:])
    for i in sList:
        if comp in i and max(max_comp) in i:
            i=i.split(",")
            return i[1],i[2]
        
bestBanc('삼성생명')
```

## 함수 predGain(year, rate)를 다음과 같이 만들어 보세요.
### 저축기간과 이율을 사용해 이득을 예측하는 함수입니다.
1. 매개변수 year는 저축년수, rate는 이율입니다. 이율은 <b>연리</b>로 계산해주세요.
2. 입력받은 year 만큼 년수가 지나면 얻을 수 있는 이득을 percentage 값으로 return 해주세요.

```python
def predGain(year,rate):
    return str(int((((100+rate)/100)**year-1)*100))+'%'

predGain(5, 2.57)
```
