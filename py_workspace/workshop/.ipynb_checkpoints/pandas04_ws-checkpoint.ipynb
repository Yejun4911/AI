##### 2019년 의 영화 평점 데이터를 가져옵니다.

```python
import time

date = pd.date_range('20190101', periods=31, freq='D')

movie_date = []
movie_name = []
movie_point = []

for today in date:
    html = "http://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date=
                                   urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, "html.parser")
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
    time.sleep(1)
```

```python
# 수집된 데이터 건수 확인
len(movie_date), len(movie_name), len(movie_point)
```

```python
movie = pd.DataFrame({'date':movie_date, 
                      'name':movie_name, 
                      'point':movie_point})
movie.head()
```

```python
# 인터넷의 문제로 미리 수집된 데이터를 이용합니다.
# 2019년 네이버 영화 평점 리스트(2019년 네이버 영화 평점 리스트.csv 파일이용)를 movie 변수로 불러옵니다.
movie=pd.read_csv('../data/2019년 네이버 영화 평점 리스트.csv')
movie.head()
```

```python
# 데이터 프레임의 컬럼 정보를 확인합니다.
movie.info()
```

```python
# date 컬럼을 datetime 타입으로 변경합니다.
movie=movie.astype({'date': 'datetime64'})
movie.info()
```

```python
# 영화별 평점 평균을 계산합니다.
movie.pivot_table(index='name',values='point',aggfunc=['mean'])
```

```python
# 평균 평점이 높은 10개 영화를 시각화합니다.
movie.pivot_table('point',index='name',aggfunc='mean',margins=True).sort_values('point',ascending=False).head(10)
```

```python
# 평균 평점이 낮은 10개 영화를 시각화합니다.
movie.pivot_table('point',index='name',aggfunc='mean',margins=True).sort_values('point').head(10)
```

```python
# 영화별 평점 건수(리뷰 요일)를 계산합니다.
movie.pivot_table('date','name',aggfunc=['count'])
```

```python
# 평점 건수(리뷰 요일)가 높은 10개 영화를 시각화합니다.
movie.pivot_table('date','name',aggfunc='count').sort_values('date',ascending=False).head(10)
```

```python
# 평점 건수(리뷰 요일)가 높은 10개 영화를 시각화합니다.
movie.pivot_table('date','name',aggfunc='count').sort_values('date').head(10)
```

```python
# 평균 평점이 높은 순서대로 영화 정보를 추출합니다.
tmp=movie.pivot_table('point',index='name',aggfunc='mean',margins=True).sort_values('point',ascending=False)
tmp.index
```

```python
# 평균 평점 1위 영화의 평점 변화를 시각화합니다.
# 평균 평점 1위 영화의 평점 변화를 시각화합니다.

movie.pivot_table('point',index='name',aggfunc='mean',margins=True).sort_values('point',ascending=False).plot()
plt.xticks(rotation=45)
plt.figure(figsize=(12,8))
plt.show()
```

```python
# 모든 영화에 대해 날짜별 평점 변화를 계산합니다.
movie.pivot_table(index='date',values=['point'],columns='name',aggfunc='mean').head(5)
```

```python
# 평점이 없는 경우 (na의 경우) 0으로 치환합니다.

movie.pivot_table(index='date',values=['point'],columns='name',aggfunc='mean').fillna(0).head(5)
```

```python
# date 컬럼에서 년, 월, 일, 요일 정보를 추출합니다.
movie['year']=movie['date'].dt.year

movie['month']=movie['date'].dt.month

movie['day']=movie['date'].dt.day

movie['wday']=movie['date'].dt.weekday

movie['week']=movie['date'].dt.weekofyear
movie.head(100)
```

```python
# 영화의 요일별 평균 평점을 계산합니다.

movie.pivot_table(index=['name','wday'],values='point',aggfunc='mean')
```

```python
# 인덱스를 재설정합니다. reset_index함수 이용
movie.pivot_table(index=['name','wday'],values='point',aggfunc='mean').reset_index()
```

```python
# 영화의 주별 평균 평점을 계산합니다.
movie.pivot_table(index=['name','week'],values='point',aggfunc='mean')
```

```python
# 인덱스를 재설정합니다. reset_index함수 이용
movie.pivot_table(index=['name','week'],values='point',aggfunc='mean').reset_index()
```

```python
# 영화별 주차를 계산합니다.
movie.pivot_table(index='name',values='week',aggfunc='max')
```
